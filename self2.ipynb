```python
def evaluate(val):
    a=''
    dat_json = graphlab.SFrame({
        'user_id':["0","0","0","1","1","1","2","2","2","3","3","3","4","4","4","5","5"],
        'item_id':["a","c","f","b","c","f","a","c","b","c","d","d","d","d","f","a","e"],
        'rating':[1,3,5,2,3,4,5,4,1,2,3,5,1,1,4,3,2]})
    dat_json = graphlab.SFrame(dat_json)
    #print dat_json.print_rows(num_rows=100,num_columns=100)
    model_dat_json = graphlab.ranking_factorization_recommender.create(dat_json, target='rating',verbose=False)
    users = dat_json['user_id'].unique()
    string = model_dat_json.recommend(users=['user_id'==('val')])
    for i in range(len(string)):
        a+="'"
        a+=(string['item_id'][i])
        a+="'"
        if i!=(len(string)-1):
            a+=","
    return a
```
